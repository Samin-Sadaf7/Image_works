```python
randbin = np.random.choice([0,255], size=(100000,), p=[1./3,2./3])
flatnpArray =np.array(randbin,  dtype=np.uint8)
print(flatnpArray.shape)
binimg = flatnpArray.reshape(200,500)
cv2_imshow(binimg)
```

```python
choices = [ i for i in range(0,256)]
randbin = np.random.choice(choices, size=(3000000,))
flatnpArray =np.array(randbin,  dtype=np.uint8)
img = flatnpArray.reshape(1000, 1000, 3)
cv2_imshow(img)
```
